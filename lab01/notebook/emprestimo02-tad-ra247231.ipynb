# Transformando Empréstimo em um TAD

Considere a resolução para múltiplos empréstimos.

```c
#include <stdio.h>

typedef struct {
    float s;
    int   n;
    float j;
    int   corrente;
    float p;
} Emprestimo;

Emprestimo novoEmprestimo(float s, int n, float j) {
    Emprestimo umEmprestimo;
    umEmprestimo.s = s;
    umEmprestimo.n = n;
    umEmprestimo.j = j;
    umEmprestimo.corrente = 1;
    umEmprestimo.p = s;
    return umEmprestimo;
}

float proximaParcela(Emprestimo *umEmprestimo) {
    float retorno = umEmprestimo->p;
    if (umEmprestimo->corrente < umEmprestimo->n)
        umEmprestimo->p = umEmprestimo->p + (umEmprestimo->p * (umEmprestimo->j/100));
    else
        umEmprestimo->p = 0;
    (umEmprestimo->corrente)++;
    return retorno;
}

int main() {
    Emprestimo emprestimo1,
               emprestimo2;
    
    emprestimo1 = novoEmprestimo(200, 5, 1);
    emprestimo2 = novoEmprestimo(500, 7, 2);
    
    float p1 = proximaParcela(&emprestimo1),
          p2 = proximaParcela(&emprestimo2);
    
    int i = 1;
    while (p1 > 0 || p2 > 0) {
        if (p1 > 0)
            printf("Emprestimo 1: parcela %d eh %3.2f\n", i, p1);
        if (p2 > 0)
            printf("Emprestimo 2: parcela %d eh %3.2f\n", i, p2);
        p1 = proximaParcela(&emprestimo1);
        p2 = proximaParcela(&emprestimo2);
        i++;
    }   
}
```

# Exercício

Modifique este código (ou um código que você mesmo tenha feito) para que a estrutura (struct) possa representar um Tipo Abstrato de Dados (TAD), em que ela contenha não apenas os valores de estado,como também operações executadas sobre eles. As operações (funções em C) também devem fazer parte do struct do mesmo modo que os campos de dados.

```c
#include <stdio.h>

typedef struct Emprestimo {
    double parcelaAtual, juros;
    int numParcelas;
    double (*proximaParcela)(struct Emprestimo);
} Emprestimo;

typedef struct {
    Emprestimo emprestimos[100];
    int numEmprestimos;
} MultiplosEmprestimos;

double proximaParcela(Emprestimo emprestimo){
    emprestimo.parcelaAtual *= (1+emprestimo.juros/100);
}

Emprestimo criaEmprestimo(double parcelaInicial, double juros, int numParcelas){
    Emprestimo emprestimo;
    emprestimo.parcelaAtual = parcelaInicial;
    emprestimo.juros = juros;
    emprestimo.numParcelas = numParcelas;
    emprestimo.proximaParcela = proximaParcela;
    return emprestimo;
}

MultiplosEmprestimos registraEmprestimos(int qtd, Emprestimo emprestimos[100]){
    int i;
    MultiplosEmprestimos mEmps;
    for (i = 0; i < 100; i++)
        mEmps.emprestimos[i] = emprestimos[i];
    mEmps.numEmprestimos = qtd;
    return mEmps;
}

void calculaParcelas(MultiplosEmprestimos mEmps){
    int i, parcelas = 0, haParcelasIncontabilizadas = 1;
    while (haParcelasIncontabilizadas){
        haParcelasIncontabilizadas = 0;
        parcelas++;
        for (i = 0; i < mEmps.numEmprestimos; i++){
            if (parcelas <= mEmps.emprestimos[i].numParcelas) {
                printf("Emprestimo %d: parcela %d eh %.2lf\n", i + 1, parcelas, mEmps.emprestimos[i].parcelaAtual);
                mEmps.emprestimos[i].parcelaAtual = mEmps.emprestimos[i].proximaParcela(mEmps.emprestimos[i]);
                haParcelasIncontabilizadas = 1;
            }
        }
    }
}


int main(){
    Emprestimo emprestimos[100] = {criaEmprestimo(200, 1, 5), criaEmprestimo(500, 2, 7)}; 
    MultiplosEmprestimos mEmps = registraEmprestimos(2, emprestimos);
    
    calculaParcelas(mEmps);
}
```
